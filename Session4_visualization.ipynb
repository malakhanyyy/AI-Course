```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
file_id = '18Aie7PXX77qZj1mFHp2YXw-gzBstu19p'
url = f"https://drive.google.com/uc?export=download&id={file_id}"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
df['bedrooms'].value_counts()
```

```python
plt.bar(df['bedrooms'].value_counts().index,df['bedrooms'].value_counts())
plt.show()
```

```python
sns.countplot(x='bedrooms',hue= 'furnishingstatus',data= df)
plt.show()
```

```python
sns.barplot(x='bedrooms',y = 'price',data= df)
```

```python
plt.hist(df['bedrooms'], bins=14)
plt.show()
```

```python
plt.scatter(df['bedrooms'],df['price'],marker='^',color='purple')
plt.title('bedroom vs price')
plt.xlabel('bedroom')
plt.ylabel('price')
plt.show()
```

```python
plt.scatter(df['area'],df['price'],marker = 'o', c=df['price'])
plt.title('area vs price')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
plt.boxplot(df['bedrooms'])
plt.show
```

```python
sns.boxplot(df['bedrooms'])
plt.show
```

```python
furnishing_counts = df['furnishingstatus'].value_counts()
colors = ['navy' , 'pink' , 'purple']
bars = plt.bar(furnishing_counts.index,furnishing_counts.values,color=colors)
plt.legend(bars,furnishing_counts.index,title = 'furnishingstatus')
plt.title('num of houses by furnishing status')
plt.xlabel('furnishing status')
plt.ylabel('number of houses')
plt.show()
```

```python
plt.pie(furnishing_counts, labels= furnishing_counts.index, autopct= '%1.2f%%')
plt.title('num of houses by furnishing status')
```

```python
sns.catplot(x='furnishingstatus' , y='price', data=df, kind='bar')
plt.show()
```

```python
sns.relplot(x='area', y='price', data=df, kind='scatter', col='bedrooms')
plt.show()
```

```python
sns.relplot(x='area', y='price', data=df, kind='line', style='furnishingstatus')
plt.show()
```
