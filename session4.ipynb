```python
import pandas as pd
dp = pd.read_csv('/content/students.csv')
dp.head(10)
dp.tail()
dp.shape
dp.info()
dp.describe()
dp.columns
dp.dtypes
dp[["Age","Name"]]
dp.iloc[0:2,0:9]      #row then col
dp.iloc[0:2,0:]      #row then col     index onlt
dp.loc[:,['Name','Age']]    #index or name
dp[dp['GPA']>2.98]     #Filtering
dp.sort_values('Salary' , ascending=True)
dp.sort_values('Salary' , ascending=False)
dp.isnull().sum()
dp['Salary']=dp['Salary'].fillna(dp['Salary'].mean())
dp['GPA']=dp['GPA'].fillna(dp['GPA'].mean())
dp.isnull().sum()
dp.duplicated().sum()
dp[dp.duplicated]
dp = dp.drop_duplicates()
dp.groupby('Department')['Salary'].agg('mean','max','min')
dp.pivot_table(values='Salary',index='Department',aggfunc='mean')
dp['Bonus']= dp['Salary'].apply(lambda x:x*4)
dp['Gender_code']= dp['Gender'].map({'Male': 0,'Female': 1})
dp[['Gender','Gender_code']].head()
dp['Passed']=dp['Passed'].replace({'Yes':'Pass','No':'Fail'})
dp[['Passed']].head()


```

```python
grades = pd.read_csv('grades.csv')
# pd.merge(dp,grades,on='ID').head()
clean_students = pd.merge(dp, grades, on='ID')
clean_students.to_csv('clean_students.csv', index=False)
clean_students.head()
!ls
```
